Data Analysis for 15 queries with sql

```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("sales_data_analysis.db")
cur = con.cursor()

# Define each query with its description and limit the results for clarity
queries = {
    "Customer Distribution by Gender": "SELECT Gender, COUNT(*) AS Count FROM sales_data GROUP BY Gender;",
    "Top 10 Products by Sales Amount": "SELECT `Product Name`, SUM(`Sales Amount`) AS Total_Sales FROM sales_data GROUP BY `Product Name` ORDER BY Total_Sales DESC LIMIT 10;",
    "Sales Trend Over Time (Top 10 Dates)": "SELECT `Order Date`, SUM(`Sales Amount`) AS Total_Sales FROM sales_data GROUP BY `Order Date` ORDER BY `Order Date` DESC LIMIT 10;",
    "Sales by Category": "SELECT Category, SUM(`Sales Amount`) AS Total_Sales FROM sales_data GROUP BY Category LIMIT 10;",
    "Top 10 Store Sizes by Total Sales": "SELECT `Square Meters`, SUM(`Sales Amount`) AS Total_Sales FROM sales_data GROUP BY `Square Meters` ORDER BY Total_Sales DESC LIMIT 10;",
    "Top 10 Customer Ages by Purchase Frequency": "SELECT Age, COUNT(*) AS Count FROM sales_data GROUP BY Age ORDER BY Count DESC LIMIT 10;",
    "Seasonality of Sales (Monthly)": "SELECT strftime('%m', `Order Date`) AS Month, SUM(`Sales Amount`) AS Total_Sales FROM sales_data GROUP BY Month ORDER BY Month;",
    "Discounted Sales by Category (Top 10 Categories)": "SELECT Category, SUM(`Sales Amount`) AS Total_Sales FROM sales_data WHERE Discounted = 1 GROUP BY Category ORDER BY Total_Sales DESC LIMIT 10;",
    "Top 10 Regions by Sales Over Time": "SELECT Country_x, SUM(`Sales Amount`) AS Total_Sales FROM sales_data GROUP BY Country_x ORDER BY Total_Sales DESC LIMIT 10;",
    "Top 10 Product Categories by Sales": "SELECT Category, SUM(`Sales Amount`) AS Total_Sales FROM sales_data GROUP BY Category ORDER BY Total_Sales DESC LIMIT 10;",
    "Sample Product Price vs. Sales (Scatter Plot Data)": "SELECT `Unit Price USD`, `Sales Amount` FROM sales_data LIMIT 15;",
    "Top 10 Loyal Customers (Repeat Purchases)": "SELECT CustomerKey, COUNT(`Order Number`) AS Repeat_Purchases FROM sales_data GROUP BY CustomerKey HAVING Repeat_Purchases > 1 ORDER BY Repeat_Purchases DESC LIMIT 10;",
    "Average Basket Size (Products per Order)": "SELECT `Order Number`, COUNT(`Product Name`) AS Products_Per_Order FROM sales_data GROUP BY `Order Number` LIMIT 10;",
    "Discounted vs. Non-Discounted Sales Performance": "SELECT Discounted, SUM(`Sales Amount`) AS Total_Sales FROM sales_data GROUP BY Discounted;",
    "Top 10 Product Return Rates": "SELECT `Product Name`, AVG(Return) AS Return_Rate FROM sales_data GROUP BY `Product Name` ORDER BY Return_Rate DESC LIMIT 10;"
}

# Execute each query and print top results in an organized manner
for description, query in queries.items():
    print(f"--- {description} ---")
    cur.execute(query)
    results = cur.fetchall()
    for row in results:
        print(row)
    print("\n" + "="*40 + "\n")  

con.close()

```
