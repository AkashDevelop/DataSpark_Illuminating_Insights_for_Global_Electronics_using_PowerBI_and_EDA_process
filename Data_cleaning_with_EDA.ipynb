# Insights from Skewness:
Positive Skewness:
Variables like Line Item (1.33), Quantity (1.11), Unit Cost USD (2.90), and Unit Price USD (3.56) have positive skewness. This indicates that these distributions have longer right tails, suggesting that there are a few instances of high values.
Negative skewness:
Variables like Order Number (-0.54), CustomerKey (-0.26), and others exhibit negative skewness. This suggests that there are lower values that might be affecting the mean, indicating a concentration of data towards higher values.

# Insights from Kurtosis:

Leptokurtic Distribution (Kurtosis > 0):
Line Item (1.48), Unit Cost USD (11.92), and Unit Price USD (17.57) have positive kurtosis, indicating a leptokurtic distribution. This suggests that the data has heavier tails and a sharper peak than a normal distribution, indicating more frequent extreme values.

Platykurtic Distribution (Kurtosis < 0):
Variables such as Order Number (-0.51), CustomerKey (-1.12), and StoreKey (-1.45) have negative kurtosis, indicating a platykurtic distribution. This means that the data is more uniform with lighter tails and a flatter peak compared to a normal distribution

```python
# Define a function to detect outliers using Tukey's method
def detect_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Detect outliers in each specified column
outliers_order_number = detect_outliers(merged_data, 'Order Number')
outliers_customer_key = detect_outliers(merged_data, 'CustomerKey')
outliers_zip_code = detect_outliers(merged_data, 'Zip Code')
outliers_square_meters = detect_outliers(merged_data, 'Square Meters')
outliers_unit_price= detect_outliers(merged_data,'Unit Price USD')

# Print the number of outliers found in each column
print(f"Outliers in Order Number: {outliers_order_number.shape[0]}")
print(f"Outliers in CustomerKey: {outliers_customer_key.shape[0]}")
print(f"Outliers in Zip Code: {outliers_zip_code.shape[0]}")
print(f"Outliers in Square Meters: {outliers_square_meters.shape[0]}")
print(f"outliers in price unit: {outliers_unit_price.shape[0]}")

# Optionally, display the outlier data
# Uncomment these lines if you want to see the outlier records
print(outliers_order_number)
print(outliers_customer_key)
print(outliers_zip_code)
print(outliers_square_meters)
print(outliers_unit_price)
```

```python
#  we can handle the outliers in the Square Meters column by using either capping
#well replace the outliers with the nearest non-outlier values based on the interquartile range (IQR) bounds. This keeps the data within a realistic range without removing any records.


# Calculate IQR
Q1 =merged_data['Square Meters'].quantile(0.25)
Q3 = merged_data['Square Meters'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Cap the outliers
merged_data['Square Meters'] = merged_data['Square Meters'].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))

```

```python
Q1 =merged_data['Unit Price USD'].quantile(0.25)
Q3 = merged_data['Unit Price USD'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Cap the outliers
merged_data['Unit Price USD'] = merged_data['Unit Price USD'].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))
```

```python
#  Circular Bar Plot of Product Category Sales
# A circular bar plot can display the sales distribution by product category in a visually appealing manner.
import numpy as np
category_sales = merged_data.groupby('Category')['Sales Amount'].sum().sort_values(ascending=False)
categories = category_sales.index
values = category_sales.values
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
values = np.concatenate((values, [values[0]]))
angles += angles[:1]

plt.figure(figsize=(10, 10))
ax = plt.subplot(111, polar=True)
ax.bar(angles, values, width=0.3, bottom=0.0)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, rotation=45)
ax.set_title('Sales Distribution by Product Category')
plt.show()
```
